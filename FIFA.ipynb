# Checking the data types for proper cleaning.

```python
FIFA.dtypes
```

# Removing the newline expression in the dataset.

```python
FIFA = FIFA.replace('\n',' ', regex=True)
```

# Filling Na Values.

```python
FIFA = FIFA.fillna('')
```

```python
FIFA
```

# Splitting the Joined columns into Day, Month, and Year.

```python
FIFA[['Month', 'Day', 'Year']] = FIFA['Joined'].str.split(' ', expand=True)
```

```python
FIFA
```

```python
FIFA = FIFA.replace(',','', regex=True)
```

```python
FIFA
```

```python
FIFA.drop_duplicates()
```

# Dropping some columns.

```python
FIFA = FIFA.drop(columns = ['Joined', 'photoUrl', 'playerUrl'])
```
